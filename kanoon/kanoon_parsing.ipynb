## 1. Creating a Citation Network
Looping over the text of each case, create a dictionary with the key as a kanoon case id and the value as a set of kanoon case ids or laws that are cited in the text. So for example, if kanoon case id 10 cites a kanoon document id number 201374 we have in our network 10: {201374}.

```python
# We loop over each file and find all citations using regex and add it to the citations dictionary with the case 
# as the key and a set of citations as the value.


start_time = time.time()

def create_network(list_of_filenames):
    citation_network = {}
    for filename in list_of_filenames:
        with open((os.path.join(head, 'sample_cases_10k', filename)), encoding="utf8") as f:
            text = f.read()
            # Use '/[a-zA-z]+/' below if you want to expand search from /doc/ to others. 
            citations = (set(re.findall("""/doc*/(\d*)""", text, flags=re.IGNORECASE)))
            citations.discard('')
            # Every case cites itself for newer cases due to website structure. Comment below line if you do want that.
            citations.discard(filename[:-4])  
            if citations:
                citation_network[filename[:-4]] = citations                
    return citation_network

citation_network = create_network(list_of_filenames)
    
print(f"--- {(time.time() - start_time)} seconds ---")  # takes 200-300 seconds
```

### Creating a Sparse Matrix and Edge DataFrame from Citation Network
The final product is a matrix of n by n dimensions where n is the number of cases we are analyzing. So for example if we run our final model on a million cases we would need a million by million matrix of 0's and 1's with 1 denoting where a case is citing another. This would be computationally impossible so we build a sparse matrix instead using scipy's sparse module. This sparse matrix of 1's and 0's could be useful in future network analysis.

```python
# First we create a sparse matrix by identifying the indexes where the value is 1 (citation present).
# Convert this sparse matrix to a sparse dataframe with index and col names as list of filenames. 

from scipy.sparse import coo_matrix

start_time = time.time()
row_indices = []
col_indices = []

for i, filename in enumerate(list_of_filenames):
    try:
        if citation_network[filename[:-4]]:
            for citation in citation_network[filename[:-4]]:
                if citation+'.txt' in list_of_filenames:
                    row_indices.append(i)
                    col_indices.append(list_of_filenames.index(citation+'.txt'))
    except KeyError:
        continue

data = np.ones(len(row_indices))

m = len(list_of_filenames)
network_matrix = coo_matrix((data, (row_indices, col_indices)), shape = (m, m))
citations_df = pd.DataFrame.sparse.from_spmatrix(network_matrix, index=list_of_filenames, columns=list_of_filenames)
# citations_df.to_csv('citations_sparse.csv')

print(f"--- {(time.time() - start_time)} seconds ---")  # takes 5-10 seconds
citations_df.head()
```

We create an edge list where each row is a (case, citation) tuple. This is helpful for visualizing in Gephi. 

```python
edge_list = [ ]

for case, citation_set in citation_network.items():
    if citation_set:
        edge_list.extend(((case, citation) for citation in citation_set if case!=citation))

df = pd.DataFrame(edge_list, columns = ['Case', 'Citation'])
df.to_csv('citation_edge_list.csv', index=False) 
df.head(5)
```

### Create Acts Network
Similar to the citation network, here we only do a subset of the citations - creating a dictionary where the key is the kanoon id of a case and the value is a set of acts/subsections cited in that case text. We have created a separate notebook called acts_analysis to explore the act-case network in further detail. Use that notebook to see the latest work in this regard.

```python
acts_network = {}
start_time = time.time()

for filename in list_of_filenames:
    with open(os.path.join(os.path.join(head, 'sample_cases_10k', filename)), encoding="utf8") as f:
        soup = BeautifulSoup(f, 'html.parser')
        citations = soup.find_all("div", "cite_title")
        for citation in citations:
            act_keywords = ['section', 'constitution', 'penal', 'act']
            title = citation.get_text().lower()
            if any(keyword in title for keyword in act_keywords):
                if title in acts_network:
                    acts_network[title].add(filename[:-4])
                else:
                    acts_network[title] = {filename[:-4]}

    
print(f"--- {(time.time() - start_time)} seconds ---")  # takes 200-300 seconds
```

```python
list_of_banks = pd.read_csv('../bank_names.csv', header=None)
list_of_banks = list_of_banks[0].tolist()
list_of_banks = ik.preprocess_company_names(list_of_banks, bank=True)
list_of_banks[:5]
```

```python
# We loop over each file and find all citations using regex and add it to the citations dictionary with the case 
# as the key and a set of citations as the value.

network = {}
titles = []
dates = []
courts = []
cnr_nums = []
case_nums = []
petitioner = []
respondent = []
petitioner_advocates = []
respondent_advocates = []
judges = []
banks = []
citations = []
count = 0

start_time = time.time()  

for filename in list_of_filenames:
    count+=1
#     if count>5000:
#         break
    with open(os.path.join(os.path.join(head, 'sample_cases_10k', filename)), encoding="utf8") as f:
        soup = BeautifulSoup(f, 'html.parser')
        text = soup.get_text().lower()
        try:
            title, date = soup.title.string.split(' on ')
            date = datetime.strptime(date, "%d %B, %Y")
            titles.append(title)
            dates.append(date)
            courts.append(soup.find("div", {"class": "docsource_main"}).string)
        except AttributeError:
            titles.append(np.nan)
            dates.append(np.nan)
            courts.append(np.nan)
        case_nums.append(ik.extract_case_num(text))
        petitioner.append(ik.extract_petitioner(text))
        respondent.append(ik.extract_respondent(text))
        pet_adv, resp_adv = ik.extract_petitioner_advocate(text)
        petitioner_advocates.append(pet_adv)
        respondent_advocates.append(resp_adv)
        cnr_nums.append(ik.extract_cnr(text))
        judges.append(ik.extract_judges(text, soup))
        banks.append(ik.extract_banks(text, list_of_banks))
        citations.append(ik.extract_citations(filename, str(soup)))
#     break
        
    
print(f"--- {(time.time() - start_time)} seconds ---")  # takes 1000-1500 seconds

# With 5000 files it takes 60-220 seconds to append to lists. 
```

```python
df = pd.DataFrame({'file':list_of_filenames[:], 'cnr_num': cnr_nums, 'title': titles, 'date': dates, 'court': courts, 
                   'case_number': case_nums, 'petitioner': petitioner, 'respondent':respondent, 
                   'petitioner_advocate': petitioner_advocates, 'respondent_advocate': respondent_advocates,
                   'judge': judges, 'banks': banks, 'citations': citations})
df['bench_size'] = df.loc[:,'judge'].dropna().apply(len)
df['num_citations'] = df.loc[:,'citations'].dropna().apply(len)
df['chief_justice'] = df.loc[:,'judge'].dropna().apply(lambda x: (any('chief' in item for item in x)))

keywords_state = ['state', 'government', 'commissioner', 'national', 'india', 'indian', 'public', 'magistrate',
                  'collector', 'department', 'district']
df['state'] = df.loc[:,'title'].dropna().apply(lambda x: (any(keyword in x.lower() for keyword in keywords_state)))

keywords_business = ['pvt', 'private', 'limited', 'ltd', 'company', 'co.', 'llp', 'llc']
df['business'] = (df.loc[:,'petitioner']+df.loc[:,'respondent']).dropna().apply(lambda x: (any(keyword in x.lower() for keyword in keywords_business)))

keywords_male = ['shri', 'mr.', 'master']

df.replace('\n', ' ', regex = True, inplace = True)
df.head()
```

```python
judges_dict = defaultdict(int)
for judge_set in df['judge']:
    if str(judge_set)!='nan':
        for judge in judge_set:
            judges_dict[judge] += 1
        
```

```python
print(len(judges_dict))        

for judge_pair in itertools.combinations(judges_dict, 2):
    score = jaccard.normalized_similarity(judge_pair[0], judge_pair[1])
    if score>0.95:
        carryover = judges_dict.pop(judge_pair[1], 0)
        judges_dict[judge_pair[0]] += carryover

print(len(judges_dict))   
```
