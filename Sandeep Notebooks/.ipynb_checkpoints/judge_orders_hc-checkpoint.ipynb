```python
def extract_table(response, judge_id, judge_name):
    soup = BeautifulSoup(response.text)
    table = soup.find_all('table')[8]
    rows = table.find_all('tr')
    df = pd.DataFrame([[y.text.strip() for y in x.find_all('td')] for x in rows[1:]],
                      columns = [x.text.strip() for x in rows[0].find_all('th')])
    df['judge_name'] = [judge_name]*df.shape[0]
    df['judge_id'] = [judge_id]*df.shape[0]
    #df = df[['Sl.No.', 'Order Number', 'Case Number',  'judge_id', 'judge_name',
    #         "Petitioner's Name", 'Order Date']]
    
    #df.to_csv(os.path.join(data_folder, 'judge_orders_HC', '_'.join(judge_id.split())+'.csv'), index=False)
    return df
```

```python
def extract_table(response, judge_id, judge_name):
    table = BeautifulSoup(response).find('table', attrs={'id':"ctl00_MainContent_gvSearch"})
    rows = table.find_all('tr')
    df = pd.DataFrame([[y.text.strip() for y in x.find_all('td')] for x in rows[1:]],
                      columns = [x.text.strip() for x in rows[0].find_all('th')])
    df['judge_name'] = [judge_name]*df.shape[0]
    df['judge_id'] = [judge_id]*df.shape[0]
    #df = df[['Sl.No.', 'Order Number', 'Case Number',  'judge_id', 'judge_name',
    #         "Petitioner's Name", 'Order Date']]
    
    #df.to_csv(os.path.join(data_folder, 'judge_orders_HC', '_'.join(judge_id.split())+'.csv'), index=False)
    return df
```
