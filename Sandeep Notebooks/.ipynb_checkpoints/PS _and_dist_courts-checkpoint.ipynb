```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
import os
```

```python
court_PS_data = []
dist_errors = []
PS_errors = []
```

```python

def get_dist_data(dist_code):
    
    cookies = {
        'PHPSESSID': 'j3ui7l1sld6170ojuinjmkpod2',
    }

    headers = {
        'Connection': 'keep-alive',
        'Cache-Control': 'max-age=0',
        'Upgrade-Insecure-Requests': '1',
        'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.92 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
        'Sec-Fetch-Site': 'none',
        'Sec-Fetch-Mode': 'navigate',
        'Sec-Fetch-User': '?1',
        'Sec-Fetch-Dest': 'document',
        'Accept-Language': 'en-US,en;q=0.9,bn;q=0.8',
    }

    params = (
        ('state', 'D'),
        ('state_cd', '8'),
        ('dist_cd', str(dist_code)),
    )

    response = requests.get('https://services.ecourts.gov.in/ecourtindia_v4_bilingual/cases/fir1.php', headers=headers, params=params, cookies=cookies)
    
    if response.status_code==200:
        options = BeautifulSoup(response.text).find('select',
                                                    attrs={'name':'court_complex_code'}).find_all('option')
        options = [(x.text, x['value']) for x in options[1:]]
        
        print('Number_ of options: %d' %len(options))

        # loop over options
        for i, option in enumerate(options):
            print('Working on %d out of %d' %(i+1, len(options)))
            sleep(3)
            court, val = option
            data = {
                'court_codeArr': val.split('@')[1].strip(),
                'state_code': '8',
                'dist_code': dist_code,
                'action_code': 'fillPoliceSt',
                'lang': ''}
            PS_response = requests.post('https://services.ecourts.gov.in/ecourtindia_v4_bilingual/cases/fir1_qry.php', 
                          headers=headers, cookies=cookies, data=data)
            if PS_response.status_code==200:
                court_PS_data.append((dist_code, court, val, PS_response.text))
            else:
                print('error in %d th option' %(i+1))
                print('Number of PS errors: %d' %len(PS_errors))
                print('Number of dist errors: %d' %len(dist_errors))
                PS_errors.append((dist_code, PS_response.status_code, option))
    else:
        print('district_code: %d' %dist_code)
        print('status_code: %d' %response.status_code)
        dist_errors.append((dist_code, response.status_code))
    print('Num of items scraped: %d' %len(court_PS_data))
    print('District code: %d' %dist_code)
    


#NB. Original query string below. It seems impossible to parse and
#reproduce query strings 100% accurately so the one below is given
#in case the reproduced version is not "correct".
# response = requests.get('https://services.ecourts.gov.in/ecourtindia_v4_bilingual/cases/fir1.php?state=D&state_cd=8&dist_cd=1', headers=headers, cookies=cookies)
```
